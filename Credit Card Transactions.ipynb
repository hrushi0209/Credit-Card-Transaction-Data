# Outliers Treatment

```python
# Fraud Analysis
# Compare 'Amount' for fraud vs. non-fraud transactions
df.boxplot(column='Amount', by='Fraud')
plt.title('Transaction Amount by Fraud Status')
plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['Amount'].quantile(0.25)
Q3 = df['Amount'].quantile(0.75)

# Calculate the IQR
IQR = Q3 - Q1

# Define the outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['Amount'] < lower_bound) | (df['Amount'] > upper_bound)]

print(f"Number of outliers: {len(outliers)}")

# Optionally, remove outliers
df_no_outliers = df[(df['Amount'] >= lower_bound) & (df['Amount'] <= upper_bound)]

```

```python
# Apply log transformation
df['log_Amount'] = np.log(df['Amount'] + 1)  # Adding 1 to avoid log(0)
```
